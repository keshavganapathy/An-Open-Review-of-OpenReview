```python
import pandas as pd
import random
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
import math
import pickle
sns.set_style({'font.family':'serif', 'font.serif':'Times New Roman'})
```

### Get Data In 

```python
df = pd.read_csv("../dataset_final.csv", quotechar = '"')

def clean_decision(row):
    if "Reject" == row['decisions'] or "Invite to Workshop Track" == row['decisions'] or "Withdrawn" == row['decisions']:
        return 0
    return 1

def calculate_average_score(row):
    ratings = [float(rating) for rating in row['ratings'].split(";")]
    return round(sum(ratings) / len(ratings), 1)

data = dict()

for year in range(2017, 2021):
    data[year] = df[df.year==year][['paper', 'ratings', 'decisions', 'categories']].copy()
    data[year]['average_score'] = data[year].apply(calculate_average_score, axis = 1)
    data[year]['decisions'] = data[year].apply(clean_decision, axis = 1)
    
data[2020].head(3)
```

### Using Logistic Regression to Predict Probability of a Paper Being Accepted Given Score

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

def get_model(data, xlabel="average_score", ylabel="decisions"):
    """Returns fitted logistic regression model on data"""
    X, y = np.array(data[xlabel]).reshape(-1, 1), np.array(data[ylabel])
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1, random_state = 0)
    LogReg = LogisticRegression()
    LogReg.fit(X_train, y_train)
    return LogReg

def accept_proba(model, score):
    """Returns the probability of a paper being accepted given score"""
    return model.predict_proba([[score]])[0][1]
```

### Using Hosmer-Lemeshow Goodness of Fit test to validate Logistic Regression

visit https://en.wikipedia.org/wiki/Hosmer%E2%80%93Lemeshow_test for more detail

```python
lm_model = get_model(data[2020])
gof_data = data[2020][['average_score', 'decisions']].copy()
gof_data['prob_=_1'] = gof_data.apply(lambda r: accept_proba(lm_model, r['average_score']), axis = 1)

# We've tried using 10 group originally but it didn't work because there are group with average probability = 0
# which makes the denominator of the test statistics be 0. Therefore, I chose 8 groups instead.
gof_data['group'] = pd.cut(gof_data['prob_=_1'], np.percentile(gof_data['prob_=_1'],np.linspace(0,100,9)),labels=False,include_lowest=True)
gof_data.head(5)
```

```python
test_statistic = 0
for g in range(gof_data['group'].max()):
    observed_events = len(gof_data[(gof_data['group'] == g) & (gof_data['decisions'] == 1)])
    mean_prob = np.mean(gof_data[gof_data['group'] == g]['prob_=_1'])
    expected_events = mean_prob * len(gof_data[gof_data['group'] == g])
    test_statistic += (observed_events - expected_events) ** 2 / (expected_events * (1 - mean_prob))
```

```python
def one_shot_gof(data, year):
    lm_model = get_model(data[year])
    gof_data = data[year][['average_score', 'decisions']].copy()
    gof_data['prob_=_1'] = gof_data.apply(lambda r: accept_proba(lm_model, r['average_score']), axis = 1)
    gof_data['group'] = pd.cut(gof_data['prob_=_1'], np.percentile(gof_data['prob_=_1'],[0, 12.5, 25, 37.5, 50, 62.5, 75, 87.5, 100]),labels=False,include_lowest=True)
    test_statistic = 0
    for g in range(gof_data['group'].max() + 1):
        observed_events = len(gof_data[(gof_data['group'] == g) & (gof_data['decisions'] == 1)])
        mean_prob = np.mean(gof_data[gof_data['group'] == g]['prob_=_1'])
        expected_events = mean_prob * len(gof_data[gof_data['group'] == g])
        test_statistic += (observed_events - expected_events) ** 2 / (expected_events * (1 - mean_prob))
    pvalue=1-chi2.cdf(test_statistic,6)
    return pvalue

for year in range(2017, 2021):
    print("P-value in", year, "is", one_shot_gof(data, year))
```

```python
def one_shot_gof_array(data, year):
    lm_model = get_model(data[year])
    gof_data = data[year][['average_score', 'decisions']].copy()
    gof_data['prob_=_1'] = gof_data.apply(lambda r: accept_proba(lm_model, r['average_score']), axis = 1)
    gof_data['group'] = pd.cut(gof_data['prob_=_1'], np.percentile(gof_data['prob_=_1'],[0, 12.5, 25, 37.5, 50, 62.5, 75, 87.5, 100]),labels=False,include_lowest=True)
    test_statistic = 0
    result = []
    for g in range(gof_data['group'].max() + 1):
        observed_events = len(gof_data[(gof_data['group'] == g) & (gof_data['decisions'] == 1)])
        mean_prob = np.mean(gof_data[gof_data['group'] == g]['prob_=_1'])
        expected_events = mean_prob * len(gof_data[gof_data['group'] == g])
        result.append((observed_events, expected_events))
    return result

# compare observed number of accepted papers vs the expected number of accepted papers from the model
observed_accepts = dict()
expected_accepts = dict()

for year in range(2017, 2021):
    observed_accepts[year] = []
    expected_accepts[year] = []
    for observed, expected in one_shot_gof_array(data, year):
        observed_accepts[year].append(int(observed))
        expected_accepts[year].append(int(expected))

# create x labels for plot
percentiles = [0, 12.5, 25, 37.5, 50, 62.5, 75, 87.5, 100]
x = []
for i in range(1, len(percentiles)):
    x.append(str(percentiles[i-1])  + "-" + str(percentiles[i]))

def plot_gof_seaborn(year, x, observed_accepts, expected_accepts, name):
    # Seaborn plot
    sns.set_context('talk')

    # create dataframe for plotting
    plot_df = pd.DataFrame(dict({'Decile': x, 'Observed': observed_accepts[year], 'Expected': expected_accepts[year]}))

    # initiailize figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    tidy = plot_df.melt(id_vars='Decile').rename(columns=str.title)
    tidy.columns = ['Decile', 'Type', 'Rate']
    barplot = sns.barplot(x='Decile', y='Rate', hue='Type', data=tidy, ax=ax)

    # Axis styling.
    
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    ax.tick_params(bottom=False, left=False)
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, color='#EEEEEE')
    ax.xaxis.grid(False)
    plt.legend(fontsize=18, title_fontsize=18)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(15) 
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(15)
    for item in ax.get_xticklabels():
        item.set_rotation(75)
        
    # Axis labeling
    ax.set_xlabel('Decile Rank by Acceptance Rate', labelpad=15, fontsize=20)
    ax.set_ylabel('Number of Accepted Papers', labelpad=15, fontsize=20)
    #ax.set_title('Validation of Logistic Regression Model for ' + str(year), pad=15)

        
    fig.tight_layout()
    #fig.savefig(name)
```

### Krippendorff's alpha

```python
import krippendorff
```

```python
# Approach 1: All reviewers are different
def build_reproducibility_data_1(df):
    play = df.copy()
    reliability_data = []
    paper_id = 0
    for row in play.iterrows():
        scores = list(map(int, row[1]['ratings'].split(";")))
        for score in scores:
            reviews = [np.nan] * len(play)
            reviews[paper_id] = score
            reliability_data.append(reviews)
        paper_id += 1
    return reliability_data

for year in range(2017, 2021):
    reliability_data = build_reproducibility_data_1(data[year])
    print(year, round(krippendorff.alpha(reliability_data, level_of_measurement='ordinal'), 3))
```

```python
# Approach 2: Every paper is rated by three different reviewers
def build_reproducibility_data_2(df):
    play = df.copy() # df[df['ratings'].str.split(";").map(lambda x: len(x)) >= 3].copy() df.copy()
    reliability_data = [[], [], []]
    paper_id = 0
    for row in play.iterrows():
        scores = list(map(int, row[1]['ratings'].split(";")))
        reviews = [np.nan] * 3
        i = 0
        for score in scores:
            if i == 3:
                break
            reviews[i] = score
            i += 1
        for i, review in enumerate(reviews):
            reliability_data[i].append(review)
    return reliability_data

for year in range(2017, 2021):
    reliability_data = build_reproducibility_data_2(data[year])
    print(year, round(krippendorff.alpha(reliability_data, level_of_measurement='ordinal'), 3))
```

### Monte-Carlo Simulation

```python
def get_ratings_lst(df):
    """Returns a list of lists of paper ratings in numeric format. Assumes ratings exist"""
    lst = []
    
    for row in df.iterrows():
        if not isinstance(row[1]['ratings'], str):
            continue
            
        ratings = []
        for rating in row[1]['ratings'].split(";"):
            ratings.append(int(rating))
            
        lst.append(ratings)
        
    return lst

def get_group_score(df, low, high):
    """Returns a subset of papers with low < average_score < high. Assumes averege_score column exists """
    result = df.copy()
    return result[(result['average_score'] > low) & (result['average_score'] <= high)]

def bootstrap_with_lm(data, n, low, high, acceptance_bar=0, iters=1000):
    """Returns an array containing all reproducibility levels"""
    
    # build logistic regression model
    decision_model = get_model(data)
    
    # calculate deviations in each score range
    deviations_of_range = dict()
    def calculate_deviations(data, low, high):
        deviations = []
        group = get_group_score(data, low, high)
        lst_of_lsts = get_ratings_lst(group)
        for lst in lst_of_lsts:
            for score in lst:
                mean = sum(lst) / len(lst)
                deviations.append(score - mean)
        return deviations

    for score in range(low + 1, high + 1):
        if score == 2:
            deviations_of_range[(score - 1, score)] = calculate_deviations(data, score - 1.1, score)
        else:
            deviations_of_range[(score - 1, score)] = calculate_deviations(data, score - 1, score)
    
    # get a list of all rating lists from data
    lst_of_lsts = get_ratings_lst(data)
    
    # calculate average score for each paper
    averages = [sum(lst) / len(lst) for lst in lst_of_lsts]
    
    def get_deviations_for(score, n):
        """Returns randomly n deviations for a paper with score that falls in a specific score range"""
        if math.ceil(score) <= 2:
            return np.random.choice(deviations_of_range[(1, 2)], n, replace=True)
        if math.floor(score) == math.ceil(score):
            return np.random.choice(deviations_of_range[(math.floor(score) - 1, math.ceil(score))], n, replace=True)
        
        return np.random.choice(deviations_of_range[(math.floor(score), math.ceil(score))], n, replace=True)
    
    def generate_decisions(true_scores):
        """Returns decisions for all papers using logistic regression model"""
        papers_reviews = []

        for true_score in true_scores:
            deviations = get_deviations_for(true_score, n)
            paper_reviews = [deviation + true_score for deviation in deviations]
            papers_reviews.append(paper_reviews)
        
        decisions = []
        for paper_reviews in papers_reviews:
            average_score = sum(paper_reviews) / len(paper_reviews)
            if np.random.random() < accept_proba(decision_model, average_score) + acceptance_bar:
                decisions.append(1)
            else:
                decisions.append(0)

        return decisions
    
    bt = []
    for i in range(iters):
        accept_again = 0
        accept_first = 0
        print("True score")
        true_scores = np.random.choice(averages, len(lst_of_lsts), replace=True)
        print("First Decision")
        first_decisions = generate_decisions(true_scores)
        print("Second Decision")
        second_decisions = generate_decisions(true_scores)
        for first, second in zip(first_decisions, second_decisions):
            accept_again += 1 if first == 1 and second == 1 else 0
            accept_first += 1 if first == 1 else 0
        bt.append(accept_again / accept_first)
        
    return bt
```
