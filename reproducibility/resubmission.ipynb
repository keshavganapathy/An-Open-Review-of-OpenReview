```python
import pandas as pd
import seaborn as sns
from datetime import datetime
import matplotlib.pyplot as plt
from textwrap import wrap
sns.set_style({'font.family':'serif', 'font.serif':'Times New Roman'})
```

```python
df = pd.read_csv("../dataset_final.csv", quotechar = '"')
```

```python
df['earliest_appearance_date']
```

```python
def calculate_average_score(row):
    ratings = [float(rating) for rating in row['ratings'].split(";")]
    return round(sum(ratings) / len(ratings), 1)

def parse_decision(row):
    if row['decisions'] == 'Invite to Workshop Track' or row['decisions'] == 'Reject' or row['decisions'] == 'Withdrawn':
        return "reject"
    return "accept"

def days_between(row):
    temp1 = row['earliest_appearance_date'].split("/")
    temp1[2] = "20" + temp1[2]
    temp1 = "/".join(temp1)
    temp2 = row['submission_date'].split("/")
    temp2[2] = "20" + temp2[2]
    temp2 = "/".join(temp2) 
    d1 = datetime.strptime(temp1, "%m/%d/%Y")
    d2 = datetime.strptime(temp2, "%m/%d/%Y")
    print(d2, d1, (d2 -d1).days)
    return (d2 - d1).days

df['average_score'] = df.apply(calculate_average_score, axis = 1)
df['days'] = df.apply(days_between, axis = 1)
df['decisions'] = df.apply(parse_decision, axis = 1)
```

```python
sum(df['days'] > 0)
```

```python
def total_papers(df, year, low, high):
    return len(df[(df.year==year) & (df.days < high) & (df.days > low)])

def accept_papers(df, year, low, high):
    return len(df[(df.year==year) & (df.days < high) & (df.days > low) & (df.decisions == "accept")])

def average_score(df, year, low, high):
    return df[(df.year==year) & (df.days < high) & (df.days > low)]['average_score'].mean()

def accept_rate(df, year, low, high):
    return accept_papers(df, year, low, high) / total_papers(df, year, low, high)
```

```python
# acceptance rate
old_accept_rates = []
middle_accept_rates = []
young_accept_rates = []

for year in range(2017, 2021):
    old_accept_rates.append(accept_rate(df, year, 180, 545))
    middle_accept_rates.append(accept_rate(df, year, 90, 180))
    young_accept_rates.append(accept_rate(df, year, float('-inf'), 90))
    
# absolute frequency of accepted papers
old_accepts = []
middle_accepts = []
young_accepts = []

for year in range(2017, 2021):
    old_accepts.append(accept_papers(df, year, 180, float('inf')))
    middle_accepts.append(accept_papers(df, year, 90, 180))
    young_accepts.append(accept_papers(df, year, float('-inf'), 90))


old_accept_scores = []
middle_accept_scores = []
young_accept_scores = []

for year in range(2017, 2021):
    old_accept_scores.append(average_score(df, year, 180, 545))
    middle_accept_scores.append(average_score(df, year, 90, 180))
    young_accept_scores.append(average_score(df, year, float('-inf'), 90))
```

```python
def plot_seaborn(x, young, middle, old, file, ylabel, ylim):
    # Seaborn plot
    sns.set_context('talk')

    # create dataframe for plotting
    plot_df = pd.DataFrame(dict({'Year': x,
                                 '< 3 months': young,
                                 '> 3 months & < 6 months': middle,
                                 '> 6 months': old}))

    # initiailize figure and axis
    fig, ax = plt.subplots(figsize=(8, 6))
    tidy = plot_df.melt(id_vars='Year').rename(columns=str.title)
    tidy.columns = ['Year', 'Age', ylabel]
    barplot = sns.barplot(x='Year', y=ylabel, hue='Age', data=tidy, ax=ax)

    # Axis styling.
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_color('#DDDDDD')
    ax.tick_params(bottom=False, left=False)
    ax.set_axisbelow(True)
    ax.yaxis.grid(True, color='#EEEEEE')
    ax.xaxis.grid(False)
    ax.set(ylim=ylim)
    plt.legend(fontsize=18, title_fontsize=18)
    for tick in ax.xaxis.get_major_ticks():
        tick.label.set_fontsize(15) 
    for tick in ax.yaxis.get_major_ticks():
        tick.label.set_fontsize(15)  
    # Axis labeling
    split_text = textwrap.wrap("Do papers that have been public on the internet for more than 3 months at the time of submission have better acceptance rate or worse ?", 
                                width=50)
    
    ax.set_xlabel("Year", labelpad=15, fontsize=20)
    ax.set_ylabel(ylabel, labelpad=15, fontsize=20)
    #ax.set_title('\n'.join(split_text), pad=15)

    fig.tight_layout()
    #fig.savefig(file)
```

```python
plot_seaborn([2017, 2018, 2019, 2020], young_accept_rates, middle_accept_rates, old_accept_rates, "temp7.pdf", "Acceptance Rate", (0, 1))
```

```python
plot_seaborn([2017, 2018, 2019, 2020], young_accept_scores, middle_accept_scores, old_accept_scores, "temp8.pdf", "Average Score", (0, 10))
```
